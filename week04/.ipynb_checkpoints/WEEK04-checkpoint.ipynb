```python
# 思考， 如何判断一段文本中是否存在“广州”
for i,j in enumerate(title):
    if j =="广":
        if title[i+1] =="州":
            print ("广州这个词在这段文本中")
```

```python
## in 检查成员对象
"广州"in title 
```
